```python
# Outlier Removal
def outliers(udata,features):
    data = udata.copy()
    for feature in features:
        lower_q = data[feature].quantile(.25)
        higher_q = data[feature].quantile(.75)
        iqr = higher_q - lower_q
        lower_q = lower_q - 1.5 * iqr
        higher_q = higher_q + 1.5 * iqr
        data.drop(data[(data[feature]<lower_q)|(data[feature]>higher_q)].index,inplace=True)
    return data
```

```python
udata = outliers(udata,['WindGustSpeed','Humidity9am','Humidity3pm','temp_extracted','pressure_extract','Max_Min_tem_extr'])
```

```python
def feature_scale(transformer=None,column_name=''):
    transformed_data = transformer.fit_transform(udata[[column_name]])
    print(skew(np.ravel(transformed_data)))
    plt.figure(figsize=(15,6),facecolor='linen')
    plt.suptitle(f'{column_name} BEFORE AND AFTER SCALING',size=20,fontdict={'fontfamily':'serif','color':'tan'})
    plt.subplot(211)
    sns.histplot(x=udata[column_name],bins=30,color='#8bd3c7',kde=True,element='step').set_facecolor('linen')
    plt.figure(figsize=(15,6),facecolor='linen')
    plt.subplot(212)
    sns.histplot(x=np.ravel(transformed_data),bins=30,color='#8bd3c7',kde=True,element='step').set_facecolor('linen')
    plt.show()
    
    
```

```python
# Custom binning function
def custom_binning(X):
    # X is passed as a DataFrame or numpy array
    X_binned = pd.cut(X.squeeze(), bins=[-1, 3, 20, 50, 470], labels=[0, 1, 2, 3])
    return X_binned.to_numpy().reshape(-1, 1)


trf = ColumnTransformer(
    [
    
        ('bin',FunctionTransformer(custom_binning,validate=False,feature_names_out='one-to-one'),[1]),
        ('ohe',OneHotEncoder(drop='first',handle_unknown='ignore',sparse_output=False),[0]),
        ('ptrf',PowerTransformer(),[2,3,4,7,8,9]),
        ('min_max',MinMaxScaler(clip=True),[5,6])
    
    ]
    
                        ,remainder='passthrough')


```

```python
class MyDataTransformation(BaseEstimator,TransformerMixin):
    def __init__(self):
        pass
    def fit(self,X,y=None):
        return self
    def transform(self,X,y=None):
        X = X[['Date','Rainfall','Location','WindGustSpeed','Humidity9am',
            'Humidity3pm','RainToday','RainTommarow',
            'MaxTemp','MinTemp','Pressure3pm','Pressure9am','Temp3pm','Temp9am']]
        
        X['Year'] = pd.to_datetime(X['Date']).dt.year
        X['Month'] = pd.to_datetime(X['Date']).dt.month
        X['Max_Min_tem_extr'] = ((X.MaxTemp)-(X.MinTemp))/2
        X['pressure_extract'] = (X.Pressure3pm) + (X.Pressure9am)
        X['temp_extracted'] = ( X.Temp3pm - X.Temp9am )
        X['Location'] = X['Location'].replace(small_count_city,'others')
        X['RainToday'] = X.RainToday.apply(lambda x:0 if x == 'No' else 1)

        X = X[['Rainfall','Location','WindGustSpeed','Humidity9am',
            'Humidity3pm','Year','Month',"Max_Min_tem_extr",'pressure_extract'
            ,'RainToday','RainTommarow','temp_extracted']]
        
        for feature in ['Rainfall','WindGustSpeed','Humidity9am','Humidity3pm','Max_Min_tem_extr','pressure_extract']:
            X[feature].ffill(inplace=True)
        return X[['Location', 'Rainfall', 'WindGustSpeed', 'Humidity9am', 'Humidity3pm',
       'Year', 'Month', 'temp_extracted', 'pressure_extract',
       'Max_Min_tem_extr', 'RainToday', 'RainTommarow']]
        
```

```python
classifier_dict = {
    "Logistic Regression": LogisticRegression,
    "Support Vector Machine": SVC,
    "Random Forest": RandomForestClassifier,
    "Gradient Boosting": GradientBoostingClassifier,
    "AdaBoost": AdaBoostClassifier,
    "Decision Tree": DecisionTreeClassifier,
    "Extra Tree": ExtraTreeClassifier,
    "K-Nearest Neighbors": KNeighborsClassifier,
    "Gaussian Naive Bayes": GaussianNB,
    "Bernoulli Naive Bayes": BernoulliNB,
}
```

```python
def model_score():
    
    for name,model in classifier_dict.items():
        pipe = Pipeline([
            ('pre',trf),
            ('model',model())
        ])
        score = np.mean(cross_val_score(pipe,X,y,cv=5,n_jobs=-1,scoring='accuracy'))
        print(f'{name} got score of {score}')
    
```
